```python
#CREATING A np ARRAY

import numpy as np
array1 = np.array([1,2,3,4,5])
print(array1)
```

```python
#IDENTIFYING THE SIZE OF THE LIST AND np list:
import sys

#IDENTIFYING IN LIST:
lst=range(100)
print("size of each element in lst:",sys.getsizeof(lst))
print("size of lst: ",sys.getsizeof(lst)*len(lst))

#IDENTIFYING IN np:
nplst=np.arange(100)
print("size of each element in nplst:", nplst.itemsize)
print("size of nplst:", nplst.size*nplst.itemsize)
```

```python
#CREATING np ARRAY OF ZEROS:

arr0 = np.zeros((2,2))
print(arr0)

#ZERO ARRAY AS INT DATA TYPE:
print(np.int16(arr0))
```

```python
#CREATING AN ARRAY OF A NUMBER:
numarr=np.ones(10)
print(numarr)
print(np.int16(numarr))
```

```python
#CREATING AN IDENTITY MATRIX:

identitymatrix = np.eye(3)
print(identitymatrix)
```

```python
#GETTING THE DATATYPE OF A npARRAY:

arr1= np.array(np.int16([2,4,6,8,10]))
print(arr1.dtype)

arr2 = np.array([1,2,3,4,5])
print(arr2.dtype)
```

```python
#USING ARITHMATIC OPERATIONS:

import numpy as np

a = np.array([1, 2, 3, 4])
b=np.array([5, 6, 7, 8])
print(a + b)
print(a-b)
print(a*b)
print(a/b)
print(a ** b)
```

```python
#SLICING:

x = np.arange (10)
print("x= ",x)
print("\n x[:5]= ",x[:5])
print("\n x[5:]= ",x[5:])
print("\n x[4:7]= ",x[4:7])
print("\n x[::2]= ",x[::2])
print("\n x[1::2]= ",x[1::2])
print("\n x[::-1]= ",x[::-1])
print("\n x[5::-2]= ",x [5::-2])
```

```python
#BOOLEAN EXPRESSION IN NUMPY

import numpy as np

a= np.arange(12).reshape(3,4)
print("a = ",a)
print("\n")
b = a > 4
print("b = ", b)
print("\n")
print("a [b]= ", a [b]) # 1d array
print("\n")
a [b] = 1
print("a= ",a)
```

```python
#OBTAINING DATAS FROM NUMPY ARRAY USING STRINGS AND BOOLEAN
names = np.array(['Bat', 'cat', 'Dog', 'Bat', 'Dog', 'cat', 'Dog'])
data= np.arange(20).reshape(5, 4)
print(names)
print("data = ", data)
print("\n")
print("names bool = ", names == 'Bat')
print("\n")
print("names == 'Bat' ", data [names == 'Bat'])
print("\n")
print("names == 'cat', 2: ", data [names == 'cat', 2:])
print("\n")
print("names == 'Dog',3: ", data [names == 'Dog', 3])
```
