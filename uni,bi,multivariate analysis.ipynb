```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path='/content/drive/MyDrive/Machinelearning/iris.csv'
```

```python
df=pd.read_csv(file_path)
print(df.head())
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['sepal_length'],y=[0]*len(df),hue=df['species'],s=100)
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='sepal_width',y='petal_length',hue='species',s=100)
plt.yticks([])
plt.xlabel("sepal length")
plt.title("Univariate Analysis of sepal length by species")
plt.legend(title="species")
plt.grid(True)
plt.show()
```

```python
sns.pairplot(df,hue='species',vars=['petal_width','sepal_length','petal_length'])
plt.suptitle("Multivariate Analysis of petal width,sepal length, and species",y=1.02)
plt.show()
```

```python
sns.pairplot(df, hue='species',vars=['sepal_width','sepal_length','petal_width'])
plt.suptitle("Multivariate Analysis of sepal width,sepal length,and by species",y=1.02)
plt.show()
```
